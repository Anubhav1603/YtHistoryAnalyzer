```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
from bs4 import BeautifulSoup
with open('history/search-history.html', 'r', encoding="utf-8") as f:
    contents = f.read()
    soup = BeautifulSoup(contents, 'lxml')
```

```python
# search history 
search_history_df = pd.DataFrame(columns=["searched for", "link", "time"])
searches = soup.find_all('div', class_='content-cell mdl-cell mdl-cell--6-col mdl-typography--body-1')
for search in searches:
    if "Searched for" in search.text:
        dct = {"searched for":' '.join(search.text.replace("Searched for", '').strip().split()[:-5]),
               "link":search.a["href"],
               "time":''.join(search.text.replace("Searched for", '').strip().split()[-5:])
              }
        # Using pandas.concat() to append a row
        new_row = pd.DataFrame(dct, index=[0])
        search_history_df = pd.concat([new_row,search_history_df.loc[:]]).reset_index(drop=True)
display(search_history_df)
```
